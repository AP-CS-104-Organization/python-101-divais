### 2. Enumerate
    It’s common when iterating over a sequence to want to keep track of the index of the
    current item. A do-it-yourself approach would look like:

```python
dictioanry={}          # Starting a blank dictionary
print(dictioanry)      # displaying above dictionary (will be blank)
dictioanry[0]='Tiger'  # making a key 0 and assigning value 'a' to it in above dictionary
print(dictioanry)      # displaying above dictionary (Now it is filled with a key-value pair --> 0:'a')
dictioanry[1]='Monkey' # again adding another  key-value pair in above dictionary (1:'b')
print(dictioanry)      # displaying above dictionary
dictioanry[2]='Zebra'  # again adding another  key-value pair 
# We  can keep on adding key-value pair in dictionary like this. But this is not he proper way.
print(dictioanry)      # Finally displaying the dictionary
```

* We have added multiple  key value pair in initial blank dictionary.
* Same above operation can be done in multiple efficinet ways. Check the codes below:-

```python
animal_list = ['Rabbit','Cat', 'Dog','Monkey','Tiger','Lion','Zebra'] # Make list of value you want to put in dictionary
animal_dictionary ={}          # Make blank initial dictionary
j = 0                          # Initializing for key starting from zero
for i in animal_list:          # Iteration over animal_lis.( 'i' will be each member of animal_list one by one through loop) 
    animal_dictionary [j]=i    # Now making key as 'j' and assigning value  to it as 'i'
    j += 1                     # this is shortcut way of writing : j=j+1. Incrementing 'j' by 1 as loop advances.
animal_dictionary              # Finally displaying the dictionary ouside the loop
```

 #### Same above operatin using enumerate can be  done as given below.
 * After  passing any list inside enumerate, we can iterate with tracking both index and value of the list.

```python
animal_list = ['Rabbit','Cat', 'Dog','Monkey','Tiger','Lion','Zebra']
animal_dictionary  = {}
for i, v in enumerate(animal_list): 
    animal_dictionary [i] = v  # Making 'i' (index of list ) as key of animal_dic and 'v'(member of list) as value.
animal_dictionary              # Finally displaying the dictionary ouside the loop
# Note: 'i' takes the value of index of animal_list as loop advances. So it starts from zero.
# 'v' takes value of animal_list as loop advances.
```

* How about this syntax below:-

```python
animal_list = ['Rabbit','Cat', 'Dog','Monkey','Tiger','Lion','Zebra']
animal_dictionary = {}
for v in animal_list: 
    animal_dic[v[0]] = v # making zeroth character of each 'v' as the key of the dictionary,
                         # i.e. first alphabet of each member of animal_list              
animal_dictionary 
```

### 4.Creating Dictionary from Sequences(List or Tuple)
##### First understanding zip function:


```python
key_list=[10,5,3,4,55]  # first Sequecne 
value_list=['Banana','Papaya','Apple','Mango','Grape'] #Second Sequence
print(zip(key_list,value_list))  # zip Returns zip oject( see printed result), which is an iterator of tuples
# It must be unpacked through loop or pass inside list or tuple to see each  member like this:
print('\n') # for leaving a line
print('Unpacking the zip object through loop :-')
for value in  zip(key_list,value_list):
    print(value)
print('\n') # for leaving a line
print('Unpacking the zip object by passing inside list :-')   
list(zip(key_list,value_list))
```

##### Note: 
* First item in each passed sequence (list/tuple) is paired together, and then the second item in each passed sequence are paired together  and so on.
* Sequence having smallest length decide the legth of resulting iterable object.

#### Now making dictionary using two sequence
* It’s common to occasionally end up with two sequences (one for keys and one for values) that you want to pair up
element-wise in a dictionary. 
* As a first cut, we might write code like this:

```python
key_list=[10,5,3,4,55]  # first Sequecne 
value_list=['Banana','Papaya','Apple','Mango','Grape'] #Second Sequence
dictionary1={}
for i,j in zip(key_list,value_list): # unpacking the each tuple of zipped iterator
    dictionary1[i]=j
dictionary1
```

#### Since a dict is essentially a collection of 2-tuples, the dict function accepts a list of 2-tuples. Following is the more elegant way of doing it:

```python
key_list=[10,5,3,4,55]
value_list=['Banana','Papaya','Apple','Mango','Grape']
dictionary1 = dict(zip(key_list,value_list))
dictionary1
```

### 5. Setdefault method in Dictionary

```python
words=['apple','bat','Cook','bar','atom','book','Cat','Crown']
by_letter={}
for i in words:
    letter=i[0]
    if letter not in by_letter:
        by_letter[letter]=[i]
    else:
        by_letter[letter].append(i)
by_letter
```

#### The setdefault dict method is for precisely this purpose. The preceding for loop can be rewritten as:

```python
words=['apple','bat','Cook','bar','atom','book','Cat','Crown']
by_letter={}
for word in words:
    letter = word[0]
    by_letter.setdefault(letter, []).append(word)
by_letter
```

#### Note:While the values of a dict can be any Python object, the keys generally have to be immutable objects like scalar types (int, float, string) or tuples (all the objects in the tuple need to be immutable, too). The technical term is hashability.

### 8. Generators

Having a consistent way to iterate over sequences, like objects in a list, 
is an important Python feature. This is accomplished by means of the iterator
protocol, a generic way to make objects iterable. For example, iterating over a dict
yields the dict keys:

```python
some_dict = {'a': 1, 'b': 2, 'c': 3}
for i in some_dict: #  iterating over a dict yields the dict keys
    print(i,some_dict[i])
```

An iterator is any object that will yield objects to the Python interpreter when used in
a context like a for loop. Most methods expecting a list or list-like object will also
accept any iterable object. This includes built-in methods such as min, max, and sum,
and type constructors like list and tuple:

```python
iter(some_dict)
```

```python
list(iter(some_dict))
```

```python
range(10)
```

```python
iter(range(10))
```

```python
for i in range(10):
    print(i)
```

#### Generator
* It is a concise way to construct a new iterable object. 
* Normal functions execute and return a single result at a time, generators return a sequence of multiple results if pased over a list or in a loop. 
* To create a generator, use the yield keyword instead of return in a function:

```python
def squares(n=10):
    print('Generating squares from 1 to {}'.format(n))
    for i in range(1, n + 1):
        yield i ** 2    # Writing yield istead of return
```

```python
gen=squares() # Calling the function
gen           # Now this is an iterable object which can only be unfolded  through loop or we have to pass it on list or tuple
```

* Unfolding it by passing into a list

```python
list(gen) 
```

* Unfolding through loop, thus iterate  the generator as given below:-

```python
gen=squares() # Calling the fucntions
for x in gen:
    print(x)
```

* Another way of making generator:-

```python
strings=['a','as','bat','car','dove','python']
ss=(val.upper() for val in strings if len(val)>2) ## Note: Parenthesis us used instead of Square bracket.
# Above syntax is same as list comprehension except for parenthesis is used instead of square bracket.Be careful!!
ss
```

```python
for i in ss:
    print(i)
```

* Generator for getting  squares of numbers from 1 to 10.  

```python
gen=(x**2 for x in range(1,11)) ## for making generator we have to use parenthesis
gen
```

```python
tuple(gen)
```

#####  itertools.cycle(iterable)
Make an iterator returning elements from the iterable. Repeats indefinitely  if break is not used.

```python
import itertools
count = 1
for i in itertools.cycle([1,2,3]):  
    if count > 9:  
        break
    else:  
        print(i, end = " ")  
        count += 1
```

Make an iterator that returns object over and over again. Runs indefinitely unless the times argument is specified. 

```python
count = 1
for i in itertools.repeat(['COVID-19 ','COVID-20'],9):  
    if count > 4:  
        break
    else:  
        print(i, end = " ")  
        count += 1
```

#### itertools.groupby(iterable, key=None)
Make an iterator that returns consecutive keys and groups from the iterable. The key is a function computing a key value for each element. If not specified or is None, key defaults to an identity function and returns the element unchanged. Generally, the iterable needs to already be sorted on the same key function.

##### Check out this beautiful example:

```python
names=['Alan','Adam','Wes','Will','Albert','Steven']
sorted(names) # Sorting the names list
```

```python
names=['Alan','Adam','Wes','Will','Albert','Steven'] #  names list gets sorted
```

```python
names=['Alan','Adam','Wes','Will','Steven']
first_letter=lambda x:x[0] 
names=sorted(names)
dict1={}
for letter,names in itertools.groupby(names,first_letter):
    dict1[letter]=list(names)
    
dict1
```

### 10. Exceptions Handling

```python
def divide(x, y):
        result = x / y
        print(result)
```

```python
divide(2,0) # this causes division by zero which results in error
```

```python
def divide(x, y):
    try:
        result = x / y
    except:                           # if there is an error in operations inside try,the expression inside except is executed.
        print("division by zero!")
    else:                             # If there is no any error inside try,this will also executed.
        print("result is", result)   
    finally:                          # This is always executed.
        print("executing finally clause")   
```

```python
divide(2,0)
```

### 13. Arbitrary Arguments, *args
If you do not know how many arguments that will be passed into your function, add a * before the parameter name in the function definition. This way the function will receive a tuple of arguments, and can access the items accordingly:

```python
# function without *args
def multiply(a,b,d): # this is a function which can take only three input as its argument
    c=a*b*d
    return c
```

```python
multiply(3,2,8)
```

```python
# What if we want to product 2 number or say more than 3 numbers like :
multiply(3,2,8,2) # this is an error
```

##### Above requirement can be fulfilled using *args as given below:-

```python
def multiply(*value):
    z = 1
    for num in value:
        z=z*num
       # z *= num
    return z
```

```python
# Now we can pass as many argument as we want
print(multiply(4))
print(multiply(2,3))
multiply(3, 5, 10, 6,89,9)
```

```python
# for adition
def add_all(*a):
    s=0
    for i  in a:
        s=s+i
    #print(s)
    return s
```

```python
add_all(3,2,7)
```

### 14. Arbitrary Keyword Arguments, **kwargs 
* The double asterisk form of  ** kwargs is used to pass a keyworded, variable-length argument  a function. 
* Like *args, **kwargs can take however many arguments you would like to supply to it. 
* **kwargs differs from *args in that you will need to assign keywords

```python
def print_values(**kwargs):
    for key, value in kwargs.items():
        print("The value of {} is {}".format(key, value))
```

```python
print_values(name_1="Rajan",name_2="Bijay",name_3="Harper",name_4="Eric",name_5="Rami",name_6="Aon")
# Note: The argument is treated like dictionary (with '=' in between  key and value instead of ':' in dictionary). 
```
