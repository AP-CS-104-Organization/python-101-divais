```python
%matplotlib notebook
import random as r
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
plt.style.use('fivethirtyeight')
x=[]
y=[]
index=count()

def animate(i):
    x.append(next(index))
    y.append(r.randint(0,5))
    plt.cla()
    plt.plot(x,y)
    
    
ani=FuncAnimation(plt.gcf(),animate,interval=1000)
ani
plt.tight_layout()
plt.show()
```
